#### In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk

from nltk.corpus import stopwords
```

```python
sr = stopwords.words('english')
print(sr)
```

```python
freqWords = dict()
response = urllib.request.urlopen('http://php.net/') 
html = response.read() 
soup = BeautifulSoup(html,"html5lib") 
text = soup.get_text(strip=True) 
tokens = [t for t in text.split()] 
clean_tokens = tokens[:] 
#print(clean_tokens)
for token in tokens: 
        if token in stopwords.words('english'): 
            clean_tokens.remove(token) #  remove the stop words
   
```

```python
def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(zip(wordlist,wordfreq))
```

```python
dictionary = wordListToFreqDict(clean_tokens)
```

```python
def sortFreqDict(freqdict):
    sortedDict = [(freqdict[key], key) for key in freqdict]
    sortedDict.sort()
    sortedDict.reverse()
    return sortedDict
```

```python
sortFreqDict(dictionary)
```
